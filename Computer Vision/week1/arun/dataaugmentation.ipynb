```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import skimage.transform as sktran
import skimage.exposure as skexp
import skimage.util as skutil
```

```python
def disp_image(image, caption='None'):
    plt.imshow(image);
    fig = plt.imshow(image)
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    if caption !='None':
        plt.title(caption)
```

```python
num_rows=4
num_cols=3
plt.figure(figsize=(10,15))
plt.subplot(num_rows, num_cols, 1)
disp_image(image, 'Original');
plt.subplot(num_rows, num_cols, 2)
disp_image(np.fliplr(image),'Horrizontal Flipped');
plt.subplot(num_rows, num_cols, 3)
disp_image(np.flipud(image),'Vertically Flipped');
plt.subplot(num_rows, num_cols, 4)
disp_image(skimage.transform.rotate(image,90),'+90 Rotation');
plt.subplot(num_rows, num_cols, 5)
disp_image(sktran.rotate(image,-90),' -90 Rotation');
plt.subplot(num_rows, num_cols, 6)
disp_image(sktran.rotate(image,60),' +60 Rotation');
plt.subplot(num_rows, num_cols, 7)
disp_image(sktran.resize(image,(100,100)),' Resized from 500X500 to 100X100');
plt.subplot(num_rows, num_cols, 8)
disp_image(sktran.rescale(image,0.1),' Rescaled to 10:1');
plt.subplot(num_rows, num_cols, 9)
crop_width=((50,100),(100,200), (0,0))
disp_image(skutil.crop(image, crop_width, copy=False),' Cropped'); 
plt.subplot(num_rows, num_cols, 10)
disp_image(skexp.adjust_gamma(image, gamma=0.25),' Brighter');
plt.subplot(num_rows, num_cols, 11)
disp_image(skexp.adjust_gamma(image, gamma=2),' Darker');
plt.subplot(num_rows, num_cols, 12)
disp_image(skutil.random_noise(image),' Noise added');
```
